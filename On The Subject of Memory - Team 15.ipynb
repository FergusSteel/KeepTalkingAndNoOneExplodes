```python
# INSERT CODE SEQUENCE HERE
code_sequence =  [[1,3,2,4,1],
 [3,1,2,4,3],
 [2,3,4,1,2],
 [2,1,4,3,1],
 [4,1,2,3,4]]

solution = {
    "stage1" : {
        "position": -1,
        "label": -1
    },
    "stage2" : {
        "position": -1,
        "label": -1
    },
    "stage3" : {
        "position": -1,
        "label": -1
    },
    "stage4" : {
        "position": -1,
        "label": -1
    },
    "stage5" : {
        "position": -1,
        "label": -1
    }
}
```

```python
# STAGE 1
def stage1(code_sequence):
    stage_1 = code_sequence[0]
    prompt = stage_1[4]

    if prompt == 1:
        solution["stage1"]["position"] = 1
        solution["stage1"]["label"] = stage_1[1]
    elif prompt == 2:
        solution["stage1"]["position"] = 1
        solution["stage1"]["label"] = stage_1[1]
    elif prompt == 3:
        solution["stage1"]["position"] = 2
        solution["stage1"]["label"] = stage_1[2]
    elif prompt == 4:
        solution["stage1"]["position"] = 3
        solution["stage1"]["label"] = stage_1[3]
        
def stage2(code_sequence):
    stage_2 = code_sequence[1]
    prompt = stage_2[4]

    if prompt == 1:
        for position, label in enumerate(stage_2):
            if label == 4:
                solution["stage2"]["position"] = position
                solution["stage2"]["label"] = label
                return
    elif prompt == 2:
        solution["stage2"]["position"] = solution["stage1"]["position"]
        solution["stage2"]["label"] = stage_2[solution["stage1"]["position"]] 
    elif prompt == 3:
        solution["stage2"]["position"] = 0
        solution["stage2"]["label"] = stage_2[0]
    elif prompt == 4:
        solution["stage2"]["position"] = solution["stage1"]["position"]
        solution["stage2"]["label"] = stage_2[solution["stage1"]["position"]] 
        
def stage3(code_sequence):
    stage_3 = code_sequence[2]
    prompt = stage_3[4]
    
    if prompt == 1:
        for position, label in enumerate(stage_3):
            if label == solution["stage2"]["label"]:
                solution["stage3"]["position"] = position
                solution["stage3"]["label"] = solution["stage2"]["label"]
                return
    elif prompt == 2:
        for position, label in enumerate(stage_3):
            if label == solution["stage1"]["label"]:
                solution["stage3"]["position"] = position
                solution["stage3"]["label"] = solution["stage1"]["label"]
                return
    elif prompt == 3:
        solution["stage3"]["position"] = 2
        solution["stage3"]["label"] = stage_3[2]
    elif prompt == 4:
        for position, label in enumerate(stage_3):
            if label == 4:
                solution["stage3"]["position"] = position
                solution["stage3"]["label"] = label
                return 
    
def stage4(code_sequence):
    stage_4 = code_sequence[3]
    prompt = stage_4[4]
    
    if prompt == 1:
        solution["stage4"]["position"] = solution["stage1"]["position"]
        solution["stage4"]["label"] = stage_4[solution["stage1"]["position"]] 
    elif prompt == 2:
        solution["stage4"]["position"] = 0
        solution["stage4"]["label"] = stage_3[0]
    elif prompt == 3:
        solution["stage4"]["position"] = solution["stage2"]["position"]
        solution["stage4"]["label"] = stage_4[solution["stage2"]["position"]] 
    elif prompt == 4:
        solution["stage4"]["position"] = solution["stage2"]["position"]
        solution["stage4"]["label"] = stage_4[solution["stage2"]["position"]] 
        
def stage5(code_sequence):
    stage_5 = code_sequence[4]
    prompt = stage_5[4]
    
    if prompt == 1:
        solution["stage5"]["position"] = solution["stage1"]["position"]
        solution["stage5"]["label"] = stage_5[solution["stage1"]["position"]] 
    elif prompt == 2:
        solution["stage5"]["position"] = solution["stage2"]["position"]
        solution["stage5"]["label"] = stage_5[solution["stage2"]["position"]]
    elif prompt == 3:
        solution["stage5"]["position"] = solution["stage3"]["position"]
        solution["stage5"]["label"] = stage_5[solution["stage3"]["position"]]
    elif prompt == 4:
        solution["stage5"]["position"] = solution["stage4"]["position"]
        solution["stage5"]["label"] = stage_5[solution["stage4"]["position"]]
```

```python
stage1(code_sequence)
stage2(code_sequence)
stage3(code_sequence)
stage4(code_sequence)
stage5(code_sequence)

final_sequence = f"{solution['stage1']['label']}{solution['stage2']['label']}{solution['stage3']['label']}{solution['stage4']['label']}{solution['stage5']['label']}"
```
